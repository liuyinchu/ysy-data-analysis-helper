# 示例 Examples

```python
# 导入必要的包
import numpy as np
import matplotlib.pyplot as plt
import ysy_plot_helper as yph
```

## 便捷快速绘图 Convenient and fast drawing

```python
x = np.linspace(0, 10, 1000)
```

```python
# yph.plot使用示例

# 单一绘图
noise = np.random.normal(0, 0.05, len(x))
yph.plot(x, noise, 'Noise', plot_title='Sigle drawing', y_label='Noise')

# 多个绘图
z1 = np.sin(x) + noise
z2 = np.cos(x) + noise
z3 = z1 + z2
z4 = z1 - z2
yph.plot(x, (z1, z2, z3, z4), ('z1', 'z2', 'z3', 'z4'), plot_title='Multiple drawings', y_label='Z')
```

```python
# 用yph.plot绘制scatter
yph.plot(x, (z1, z2, z3, z4), ('z1', 'z2', 'z3', 'z4'), plot_title='Multiple drawings of scatter', y_label='Z', plot_type='scatter')
```

```python
# 展示绘制数据点的功能
time = np.linspace(2, 4, 10)
displacement = time**2
velocity = 2 * time
time_fine = np.linspace(2, 4, 1000)
displace_fine = time_fine**2

yph.plot(
    time_fine, displace_fine, 'Displacement Curve', 
    plot_title='Displacement-Time Relationship with $a=2m/s^2$', 
    data_point=(time, displacement), 
    x_label='Time [s]', 
    y_label='Displacement [m]'
)
```

## 临时样式加载器 Temporary Style Loader

```python
# 打印内置绘图风格
yph.print_preset_styles()
```

```python
# 示例数据
def model(x, p): 
    return x ** (2 * p + 1) / (1 + x ** (2 * p))
x = np.linspace(0.75, 1.25, 201)
```

三种推荐的绘图风格

```python
# Recommended Loading Format
with yph.temp_style(["ysy_academic", "sky"]):
    fig, ax = plt.subplots()
    for p in [10, 15, 20, 25, 30, 40, 50, 100]:      
        ax.plot(x, model(x, p), label=p)
    ax.legend(title='Order')
    ax.set(xlabel='Voltage (mV)')
    ax.set(ylabel='Current ($\mu$A)')
    ax.autoscale(tight=True)
    plt.show()
```

```python
# science + science_color
with yph.temp_style(["science", "science_color"]):
    fig, ax = plt.subplots(dpi=300)
    for p in [10, 15, 20, 25, 40, 50, 100]:      
        ax.plot(x, model(x, p), label=p)
    ax.legend(title='Order')
    ax.set(xlabel='Voltage (mV)')
    ax.set(ylabel='Current ($\mu$A)')
    ax.autoscale(tight=True)
    plt.show()
```
